```python
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv("bottle.csv")
print(df)

df=df.dropna(axis=1,thresh=800000)
print(df)
df["Salnty"].isnull().sum()

df["T_degC"].isnull().sum()

mean_value=df["T_degC"].mean()
df["T_degC"].fillna(value=mean_value, inplace=True)

mean_value=df["Salnty"].mean()
df["Salnty"].fillna(value=mean_value, inplace=True)

df.corr()

```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
from sklearn.model_selection import train_test_split
x=df[["T_degC"]]
y=df['Salnty']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
model=LinearRegression()
model.fit(x_train,y_train)
y_prediction=model.predict(x_test)
print(y_prediction)
import matplotlib.pyplot as plt
p=plt.plot(y_test,y_prediction)
print(p)
```
